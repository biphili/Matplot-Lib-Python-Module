```python
import numpy as np
x=np.linspace(0,10,20)
x
```

```python
y1=x**2
y1
```

```python
plt.plot(x,y1,'bo-')
plt.show()
```

# Giving Color ,Line width and Marker size to Graph 

```python
plt.plot(x,y1,'bo-',linewidth=5,markersize=10)
plt.show()
```

```python
plt.plot(x,y2,'gs-',linewidth=2,markersize=4)
plt.show()
```
